```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('data/tmdb_5000_movies.csv')
credits = pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
credits.info()
```

```python
movies.info()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
def convert3(obj):
    l = []
    counter = 3
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        counter -= 1
        if counter == 0:
            break
    return l
```

```python
def extract_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['crew'] = movies['crew'].apply(extract_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
# Get a movie name to find recommendations
def recommend(movie, recom_count=5):
    movie_index = new_df[new_df['title'] == movie].index[0]
    similar_movies_indexes = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x: x[1])[1:1+recom_count]
    for [idx, _] in similar_movies_indexes:
        print(new_df['title'][idx])
```

```python
recommend('Avatar')
```
